```python
num = int ( input(" ENTER A NUMBER \n"))
flag = 0
for i in range ( 2,num):
  if( num%i ==0):
    flag = 1
    break

if flag ==0 and num>1:
  print( " PRIME NUMBER")
else:
  print( "NOT A PRIME NUMBER")

```

```python
num = int ( input( " ENTER A NUMBER \n"))
rev = 0
while num >0 :
  digit = num %10
  rev=rev*10+digit
  num = num //10
print("REVERSED NUMBER = ", rev )
```
