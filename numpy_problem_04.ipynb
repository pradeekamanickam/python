# Creating a 2D Pattern Matrix

Problem: Create a 5x5 matrix filled with 1s on the border (first row, last row, first column, last column) and 0s on the inside.

```python
import numpy as np

# Create a 5x5 matrix of ones
matrix = np.ones((5, 5))

# Set the inner part (rows 1-3, cols 1-3) to zero
matrix[1:-1, 1:-1] = 0

print("Bordered Matrix:\n", matrix)
```
