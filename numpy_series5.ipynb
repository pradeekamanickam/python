# arr.sum(arr1) - gives the sum of that array1 values  
OR
# arr1.sum()

```python
print(a.sum())
print(np.sum(a))

print(b.sum())
print(np.sum(b))

print(c.sum())
print(np.sum(c))
```

## arr.sum(arr1, axis = 1) - gives the sum of values present in all rows in that array1 values, { 1 means ROWS}

## arr.sum(arr1, axis = 0) - gives the sum of values present in all coloumns in that array1 values, { 0 means COLOUMNS}


```python
print(a)
print(np.sum(a,axis=1))
a.sum(axis=1)
```

```python
print(a)
print(np.sum(a,axis=0))
a.sum(axis=0)
```
