# **NUMPY RANDOM OPERATIONS**

```python
import numpy as np
```

```python
np.random.random(1)
```

```python
np.random.random((1,1))
```

```python
np.random.random((2,4))
```

```python
np.random.random((3,3))
```

```python
np.random.randint(1,10) #we get a random integer with in the given values
```

```python
np.random.randint(1,10,(2,2)) #we get a random integer with in the given values and the (2,2) represent the size of that array
```

```python
np.random.randint(1,10,(3,3)) # represent 3by3 array
```

```python
np.random.randint(1,10,(3,1,1)) #represent random integer & the (3,3,3) is((no.of arrays),(no.of rows),(no.of coloumns))
```

```python
np.random.rand(3,3) #represent random float & the (3,3) is the size of that array
```

```python
np.random.randn(3,3) #represent random negative float val  & the (3,3) is the size of that array
```

```python
a=np.arange(1,11,1)
print(a)
```

```python
print(np.random.choice(a))
```
