# PANDAS SERIES WITH PYTHON LIST

```python
lst = [1,2,3,4.5]
print(lst)
```

```python
series = pd.Series(lst)
print(series)
print(type(series))
```

```python
empty = pd.Series ([])

empty
```

NAMING OUR OWN INDEX.


```python
a= pd.Series(['p','q','r', 's','t'],index=[10,120,12,12,14])
a
```

```python
a= pd.Series(['p','q','r', 's','t'],index=[10,120,12,12,14],name='alphabets')
a
```

```python
scalar_series=pd.Series(10)
scalar_series
```

```python
scalar_series=pd.Series(10,index =[1,2,3,4,5])
scalar_series
```
