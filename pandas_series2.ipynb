# DATA FRAME USING LIST

```python
lst=[1,2,3,4,5]
df=pd.DataFrame(lst)
print(df)
```

```python
lst=[[1,2,3,4,5],[11,12,13,14,15],[21,22,23,24,25]]
df=pd.DataFrame(lst)
print(df)
```

```python
a=[{ 'a':5,'b':6,'c':7,'d':8},{ 'a':2,'b':14,'c':70,'d':50}]                           #DICTIOMARY KEYS REPRESENTS COLUMN NAMES
df =pd.DataFrame(a)
print(df)
```

```python
a={ 'a':5,'b':6,'c':7,'d':8},{ 'a':2,'b':14,'c':70,'d':50}                         #DICTIOMARY KEYS REPRESENTS COLUMN NAMES
df =pd.DataFrame(a)
print(df)
```

```python
b={ 'Roll no': pd.Series([1,2,3,4,5]),'NAME':pd.Series(['PRIYA','PRADEE','RITHA','SASI','MASAKA']),'MARKS':pd.Series([100,90,95,10,50])}                           #DICTIOMARY KEYS REPRESENTS COLUMN NAMES
df =pd.DataFrame(b)
print(df)
```
