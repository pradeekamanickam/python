# STUDENTS MARKS ANALYSIS USING TUPLE

```python
marks= tuple( map( int, input( "ENTER MARKS\n").split()))
flag=0
for m in marks:
  if m< 40 :
    flag = 1
    break

if flag == 0:
  print( " STUDENT PASSED ALL SUBJECTS")
else:
  print( " STUDENT FAILED IN ONE OR MORE SUBJECTS")
```

# SHOPPING CART VALIDATION USING LIST

```python
stock = set(input("ENTER THE STOCK \n").split())
cart=input ( "ENTER THE CART ITEMS : \n").split()

for item in cart:
    if (item) not in stock:
        print(item,"IS NOT AVAILABLE")
    else:
        print(item,"IS AVAILABLE")
```
