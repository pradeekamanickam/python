# **MANIPULATING DATA USING DICTIONARIES**

 1) CALCULATE TOTAL MARKS OF A STUDENT USING DICTIONARY VALUES

```python
marks={"Maths":80,"Science":75,"English":70}
total=0
for m in marks.values():
   total+=m
print("Total Marks =",total)
```

2. Identify Subject with Highest Marks Using Dictionary

```python
marks={"Maths":80,"Science":90,"English":70}
highest=max(marks,key=marks.get)
print("Highest Marks Subject =",highest)
```

3. Count Number of Products in Each Category

```python
products={"Pen":"Stationery","Book":"Stationery","Soap":"Grocery","Rice":"Grocery"}
count={}
for cat in products.values():

          count[cat]=count.get(cat,0)+1
print(count)
```

4. Calculate Average Temperature Using Dictionary

```python
temp={"Mon":30,"Tue":32,"Wed":31,"Thu":33,"Fri":30}
avg=sum(temp.values())/len(temp)
print("Average Temperature =",avg)
```

5. Identify Absent Students Using Dictionary Status

```python
attendance={"Ravi":"Present","Anita":"Absent","Kumar":"Present","Meena":"Absent"}
for name,status in attendance.items():
    if status=="Absent":
        print(name)
```

6. CAALCULATE ELECTRICITY  BILL FOR MULTIPLE CONSUMERS

```python
units = { " Ravi ": 120, " Anitha ": 200, " kumar ": 150}
rate= 5
for name , u in units.items():
  print(name,"BILL = ",u*rate )
```

7. FIND EMPLOYEES ELIGIBLE FOR BONUS USING DICTIONARY.

```python
salary = { " ravi" : 28000, "anitha":35000 ," kumar " : 22000}
for name,amt in salary.items():
  if amt >30000:
    print(name, " ELIGIBLE FOR BONUS")
```

8. Calculate Total Sales Using Dictionary

```python
sales={"Pen":500,"Book":1200,"Bag":2000}
total=sum(sales.values())
print("Total Sales =",total)
```

9. Convert Marks to Grade Using Dictionary Mapping

```python
grade={"A":80,"B":60,"C":40}
mark=int(input("Enter marks\n"))
if mark>=grade["A"]:
 print("Grade A")
elif mark>=grade["B"]:
 print("Grade B")
elif mark>=grade["C"]:
 print("Grade C")
else:
 print("Fail")
```
