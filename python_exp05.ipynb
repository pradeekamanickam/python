Handling Exceptions and String Operations in Python

1. Validate User Age Input Using Exception Handling

```python
try:
   age=int(input("Enter age\n"))
   print("Valid Age:",age)
except ValueError:
   print("Invalid input! Age must be a number")
```

2. Division Operation with Zero Division Handling

```python
try:
  a=int(input("Enter numerator\n"))
  b=int(input("Enter denominator\n"))
  print("Result =",a/b)
except ZeroDivisionError:
  print("Error: Division by zero is not allowed")
```

3. Username Length Validation Using String Operation

```python
username=input("Enter username\n")
if len(username)>=6:
  print("Valid Username")
else:
  print("Username too short")
```

4. Check Whether an Email ID Contains ‘@’ Symbol

```python
email=input("Enter email ID\n")
if "@" in email:
  print("Valid Email ID")
else:
  print("Invalid Email ID")
```

5. Password Strength Check Using String Methods

```python
pwd=input("Enter password\n")
if len(pwd)>=8 and any(ch.isdigit() for ch in pwd):
  print("Strong Password")
else:
  print("Weak Password")
```

6. File Name Extension Check Using String Split



```python
fname=input("Enter file name\n")
ext=fname.split(".")[-1]
print("File Extension =",ext)
```

7. Count Words in a Sentence Using String Methods

```python
sentence=input("Enter sentence\n")
words=sentence.split()
print("Number of words =",len(words))
```

8. ATM Withdrawal Validation Using Exception Handling

```python
balance=5000
try:
 amt=int(input("Enter withdrawal amount\n"))
 if amt<=balance:
  print("Transaction Successful")
 else:
  print("Insufficient Balance")
except ValueError:
  print("Invalid amount entered")
```

9. Replace Spaces with Hyphen in a String

```python
text=input("Enter text\n")
print("Modified Text =",text.replace(" ","-"))
```

Handling
