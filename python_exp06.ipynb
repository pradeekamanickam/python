1. Electricity Bill Calculation Using User-Defined Function

```python
def calculate_bill(units):
     if units<=100:
       return units*2
     else:
       return 100*2 + (units-100)*3
units=int(input("Enter units consumed\n"))
print("Total Bill Amount = ₹",calculate_bill(units))
```

2. Find Maximum of Three Numbers Using Function

Factorial of a number using Recursive function

```python
def fact(n):
    if  n==0:
      return 1
    else:
      return n*fact(n-1)
num=int(input("Enter a number\n"))
print("Factorial of",num,"is",fact(num))
```

Sum of digits using Recursive function.

```python
def sum_digits(n):
  if n==0:
    return 0
  else:
    return n%10 + sum_digits(n//10)
num=int(input("Enter a number\n"))
print("Sum of digits =",sum_digits(num))
```

5. Count Vowels in a String Using User-Defined Function

```python
def count_vowels(text):
     count=0
     for char in text.lower():
          if ch in "aeiou":
            count+=1
     return count
s=input("Enter a string\n")
print("Number of vowels =",count_vowels(s))
```

6. Fibonacci Series Using Recursive Function

```python
def fib(n):
  if n<=1:
    return n
  else:
    return fib(n-1)+fib(n-2)
for i in range(10):
  print(fib(i),end=" ")
```

Reverse a number using Recursive function

```python
def reverse_num(n,rev=0):
  if n==0:
    return rev
  else:
    return reverse_num(n//10,rev*10+n%10)
num=int(input("Enter a number\n"))
print("Reversed number =",reverse_num(num))
```
