# Performing file operations using pythin modules.

1.Student attendence summary

```python
with open("attendence.txt", "r") as f:
    count = 0
    for line in f:
        if "present" in line:
            count += 1

print("Total present days:", count)
```

2.Log file error detection:

```python
with open("system.log","r") as f:
  for line in f:
    if "ERROR" in line:
      print(line.strip())
```

4. Student Marks Report File Creation

```python
marks = [78, 85, 69, 90]
with open("marks.txt","w") as f:
   for m in marks:
      f.write(str(m)+"\n")
avg = sum(marks)/len(marks)
print("Average marks:", avg)
```

5. Word Count from File

```python
with open("data.txt","r") as f:
   words = f.read().split()
print("Word Count:", len(words))
```

6. File Backup Creation

```python
with open("original.txt","r") as f1, open("backup.txt","w") as f2:
f2.write(f1.read())
```

7. Remove Blank Lines from File

```python
with open("input.txt","r") as f:
   lines = [line for line in f if line.strip()]
with open("clean.txt","w") as f:
   f.writelines(lines)
```

8.Reverse a file content

```python
with open("text.txt","r") as f:
   lines = f.readlines()
with open("reverse.txt","w") as f:
   for line in reversed(lines):
    f.write(line)
```
